```python
# Convertir las columnas relevantes a tipo numérico
columnas_numericas = ['Hasta 512 kbps','Entre 512 Kbps y 1 Mbps', 'Entre 1 Mbps y 6 Mbps',
                      'Entre 6 Mbps y 10 Mbps', 'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps',
                      'Más de 30 Mbps', 'OTROS','Total']

# Convertir todas las columnas a cadenas
df10[columnas_numericas] = df10[columnas_numericas].astype(str)

# Eliminar puntos y convertir a float
for columna in columnas_numericas:
    df10[columna] = df10[columna].str.replace('.', '', regex=False).astype(float)
```

```python
# Verificar para cada fila si la columna 'Total' es la suma de las anteriores columnas
columnas_verificar = ['Hasta 512 kbps', 'Entre 512 Kbps y 1 Mbps', 'Entre 1 Mbps y 6 Mbps',
                      'Entre 6 Mbps y 10 Mbps', 'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps',
                      'Más de 30 Mbps', 'OTROS']

# Calcular la suma de las columnas anteriores para cada fila
suma_columnas_anteriores = df10[columnas_verificar].sum(axis=1)

# Verificar para cada fila si la columna 'Total' es igual a la suma de las anteriores columnas
df10['TotalCalculado'] = suma_columnas_anteriores
df10['Verificacion'] = df10['Total'] == suma_columnas_anteriores


# Imprimir el DataFrame con la columna 'Verificacion'
df10
```

```python
# Crea la columna 'Año_Trimestre' combinando 'Año' y 'Trimestre'
df10['Año_Trimestre'] = df10['Año'].astype(str) + ' TM' + df10['Trimestre'].astype(str)

# Ordena el DataFrame por 'Año' y 'Trimestre'
df10 = df10.sort_values(by=['Año', 'Trimestre'], ascending=[True, True])

plt.figure(figsize=(12, 8))
categorias = columnas_verificar

# Utiliza label en lugar de labels
for columna in columnas_verificar:
    plt.plot(df10['Año_Trimestre'], df10[columna], label=columna, alpha=0.8)

plt.xlabel('Año Trimestre', fontsize=14)
plt.ylabel('Cantidad de Conexiones', fontsize=14)
plt.legend(fontsize=10)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Ajusta la rotación y alineación de las etiquetas
plt.yticks(fontsize=10)  # Ajusta el tamaño de la fuente en el eje y
plt.title('Cantidad de Conexiones por Trimestre', fontsize=16)
plt.tight_layout()  # Ajusta automáticamente el diseño para evitar recorte
plt.show()
```
